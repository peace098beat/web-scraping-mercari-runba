```python
import re
df_models = pd.read_csv("./data/imac_model_map_filled.csv")

model_a = df_models["ModelA"].unique().tolist()
model_b = df_models["ModelB"].unique().tolist()

model_a = [s.strip() for s in model_a]
model_b = [s.strip() for s in model_b]
model_a.remove("-")

print("Model A", model_a)
print("Model B", model_b)

def parse_cpu_name(x):
    cpu_names = ["i3", "i5", "i7", "M1", "M2", "PowerPC", "Duo"]

    x = x.lower()
    for cpu_name in cpu_names:
        if cpu_name.lower() in x.lower():
            return cpu_name

def parse_cpu_clock(x):
    x = x.lower()
    s = re.findall(r"\d+(?:\.\d+)?ghz", x.lower())
    s = [float(s_.replace("ghz", "")) for s_ in s ]
    if len(s) == 0:
        return None
    elif len(s) == 1:
        return s[0]#.replace("ghz", "")
    else:
        # return " ".join(s).replace("ghz", "")
        return s

df_models["cpu_clock"] = df_models["CPU"].apply(parse_cpu_clock)
df_models["cpu_name"] = df_models["CPU"].apply(parse_cpu_name)


df_models.sample(20)
```
